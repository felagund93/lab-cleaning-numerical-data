4. Rename the columns so they follow the _PE8_ (snake case).

```python
def columns_to_snake_case(df):
    import re
    df.columns = [re.sub(' ', '_', x).lower() for x in df.columns]
    return df
```

```python
customer_df=columns_to_snake_case(customer_df)
customer_df
```

7. Define a function that differentiates between continuous and discrete variables. **Hint**: A number of unique values might be useful. Store continuous data into a `continuous` variable and do the same for `discrete` and categorical.

```python
#Find the ratio of number of unique values to the total number of unique values.

likely_category = {}
for var in customer_df.columns:
    if customer_df[var].dtype == object:
        likely_category[var]='categorical'
    elif 1.*customer_df[var].nunique()/customer_df[var].count() <= 0.05: #or some other threshold
        likely_category[var] = 'discrete'
    else:
        likely_category[var] = 'continuous'
likely_category = pd.DataFrame.from_dict(likely_category, orient='index', columns=['likely_cat'])
display(likely_category)
```

```python
def likely_cat(df):
    likely_category = {}
    for var in df.columns:
        if df[var].dtype == object:
            likely_category[var]='categorical'
        elif 1.*df[var].nunique()/df[var].count() <= 0.05: #or some other threshold
            likely_category[var] = 'discrete'
        else:
            likely_category[var] = 'continuous'
    likely_category = pd.DataFrame.from_dict(likely_category, orient='index', columns=['likely_cat'])
    return likely_category
```

```python
data_cat=likely_cat(customer_df)
data_cat
```
